```python
Comments = []#empty list
Comments_upvote = []#empty list
Time = []#empty list

try:
    comments = driver.find_elements_by_xpath('//*[@id="content-text"]')
    for i in comments:
        Comments.append(i.text) #Scraping Comments
except:
    Comments.append('-')
    time.sleep(3)
try:
    comments_upvote = driver.find_elements_by_xpath('//*[@id="vote-count-middle"]')
    for i in comments_upvote:
        Comments_upvote.append(i.text) #Scraping Comments Upvote
except:
    Comments_upvote.append('-')
    time.sleep(3)
try:
    time = driver.find_elements_by_xpath('//*[@id="header-author"]/yt-formatted-string/a')
    for i in time:
        Time.append(i.text.replace('(edited)','')) #Scrape posted time
except:
    Time.append('-')
```

```python
Hostel_url = []#empty list
Name = []#empty list
Distance = []#empty list
Rating = []#empty list
Reviews = []#empty list
PrivatesFromPrice = []#empty list
DormsFromPrice = []#empty list
Facilities = []#empty list
Description = []#empty list

url = driver.find_elements_by_xpath('//h2[@class="title title-6"]//a')
for i in url:
    Hostel_url.append(i.get_attribute('href')) #Scrape url's
time.sleep(2)

name = driver.find_elements_by_xpath('//h2[@class="title title-6"]')
for i in name:
    Name.append(i.text) #Scrape hostel name
time.sleep(2)

distance  = driver.find_elements_by_xpath('//a[@class="show-on-map"]//span[1]')
for i in distance:
    Distance.append(i.text) #Scrape Distance
time.sleep(2)

try:
    rating = driver.find_elements_by_xpath('div[@class="score orange big"]')
    for i in  rating:
        Rating.append(i.text) #Scrape Ratings
try:
    rating1 = driver.find_elements_by_xpath('//div[@class="score gray big"]')
    for i in rating1:
        Rating.append(i.text)
except:
    Rating.append('-')
time.sleep(2)

reviews = driver.find_elements_by_xpath('//div[@class="reviews"]')
for i in reviews:
    Reviews.append(i.text.replace('Total Reviews','').strip()) #Scrape Reviews
time.sleep(2)

Info = []
info = driver.find_elements_by_xpath('//div[@class="price-col"]')
for i in info:
    Info.append(i.text.replace('Dorms From\n','').replace('Privates From\n',''))
time.sleep(2)
for i in range(0,len(Info),2):
    PrivatesFromPrice.append(Info[i])
time.sleep(2)

for i in range(1,len(info),2):
    DormsFromPrice.append(Info[i]) #Scraping Price
time.sleep(2)

try:
    facilities = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
    for i in facilities:
        Facilities.append(i.text.replace('\n',',')) #Scraping Facilities
except:
    Facilities.append('-')
time.sleep(2)  

for i in Hostel_url:
    driver.get(i)
    try:
        desc = driver.find_elements_by_xpath('//*[@id="__layout"]/div/div[1]/section/div[6]/div/div[2]/div/div/div[2]')
        Description.append(i.text)#Scraping Description
    except:
        Description.append('-')
        time.sleep(2)
    
```
