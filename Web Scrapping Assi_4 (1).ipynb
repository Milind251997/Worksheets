# Scrape the details of most viewed videos on YouTube from Wikipedia:
## Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
## You need to find following details:
### A) Rank
### B) Name
### C) Artist
### D) Upload date
### E) View

```python
#importing exceptions
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
```

```python
#Setting up chrome web-driver
driver = webdriver.Chrome('chromedriver.exe')
```

```python
#getting page 
page1 = driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#scraping the most viewed youtube videos information

Name=[] #empty list
Rank = []#empty list
Artist = []#empty list
Date = []#empty list
Views = []#empty list

try:

    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")

    for i in names:

        Name.append(i.text) #scraping name of the videos

except NoSuchElementException:#handling no such element exception

    Name.append('-')

except StaleElementReferenceException:#handling Stale element exception

    Name.append('-')
time.sleep(3)  
try:

    ranks=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")

    for i in ranks:

        Rank.append(i.text) #scraping anks

except NoSuchElementException:#handling no such element exception

    Rank.append('-')

except StaleElementReferenceException:#handling Stale element exception

    Rank.append('-')
time.sleep(3)
try:
    artist = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    
    for i in artist:
        Artist.append(i.text)#scraping artist names
except NoSuchElementException:
    Artist.append('-')
except StaleElementReferenceException:#handling Stale element exception
    Rank.append('-')
time.sleep(3)

try:
    date = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    
    for i in date:
        Date.append(i.text)#scraping date when the video uploaded
except NoSuchElementException:
    Date.append('-')
except StaleElementReferenceException:#handling Stale element exception
    Date.append('-')
time.sleep(3)

try:
    view = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    
    for i in view:
        Views.append(i.text)#scraping the views
except NoSuchElementException:
    Views.append('-')
except StaleElementReferenceException:#handling Stale element exception
    Views.append('-')
```

```python
dfy = pd.DataFrame()#Making dataframe

dfy['Rank'] = Rank
dfy['Name']= Name
dfy['Artist'] =  Artist
dfy['Date'] =  Date
dfy['Views'] =  Views

dfy
```

```python
Name=[]#empty list
Type = []#empty list
Task = []#empty list
Atype = []#empty list
Instance = []#empty list
Attr = []#empty list
Year = []#empty list

try:
    name = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]')
    for i in name:
        Name.append(i.text)#scraping dataset name
except NoSuchElementException:#handling no such element exception
    Name.append('-')
time.sleep(2)

try:
    types = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]')
    for i in types:
        Type.append(i.text)#scraping type of dataset
except NoSuchElementException:#handling no such element exception
    Type.append('-')
time.sleep(2)

try:
    task = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]')
    for i in task:
        Task.append(i.text)#scraping task of the dataset
except NoSuchElementException:#handling no such element exception
    Task.append('-')
time.sleep(2)

try:
    atype = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]')
    for i in atype:
        Atype.append(i.text)#scrape attribute type of thhe data
except NoSuchElementException:#handling no such element exception
    Atype.append('-')
time.sleep(2)

try:
    instance = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]')
    for i in instance:
        Instance.append(i.text)#scrape No. of Instance
except NoSuchElementException:#handling no such element exception
    Instance.append('-')
time.sleep(2)

try:
    attr = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]')
    for i in attr:
        Attr.append(i.text)#scrape No.of attribute
except NoSuchElementException:#handling no such element exception
    Attr.append('-')
time.sleep(2)

try:
    year = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]')
    for i in year:
        Year.append(i.text)#scraping year
except NoSuchElementException:#handling no such element exception
    Year.append('-')
```

```python
dfd = pd.DataFrame()#Making DataFrame

dfd['Dataset Name']= Name 
dfd['Data Type'] = Type 
dfd['Task'] = Task 
dfd['Attribute type'] = Atype 
dfd['No. of Instance'] = Instance  
dfd['No of Attribute'] = Attr 
dfd['Year'] = Year 

dfd
```
